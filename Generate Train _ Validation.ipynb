```python
df = pd.read_csv("kaggle_new_clean.csv")
```

```python
lyricsList = df["lyrics"].tolist()
```

```python
train, test = train_test_split(lyricsList, train_size=.9995, random_state=2021)
```

```python
train, eval = train_test_split(train, train_size=.95, random_state=2021)
```

```python
print(f"train: {len(train)}")
print(f"test: {len(test)}")
print(f"eval: {len(eval)}")
```

```python
dftrain = pd.DataFrame(train)
dftrain = dftrain.rename(columns={0: "lyrics"})
```

```python
dfeval = pd.DataFrame(eval)
dfeval = dfeval.rename(columns={0: "lyrics"})
```

```python
dftest = pd.DataFrame(test)
dftest = dftest.rename(columns={0: "lyrics"})
```

```python
dftrain.to_csv("/content/drive/My Drive/Datasets/all_genres/kaggle_train.csv")
dfeval.to_csv("/content/drive/My Drive/Datasets/all_genres/kaggle_validation.csv")
dftest.to_csv("/content/drive/My Drive/Datasets/all_genres/kaggle_test.csv")
```
